### Configurações basicas

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
```

```python
import os
# Entrar na página de login
caminho = os.getcwd()
arquivo = caminho + r"\login.html"
navegador.get(arquivo)
```

```python
#Preencher login, senha 
navegador.find_element(By.XPATH, '/html/body/div/form/input[1]').send_keys("Danilo@teste.com")
navegador.find_element(By.XPATH, '/html/body/div/form/input[2]').send_keys("admin")
#Clicar no botao de fazer login
navegador.find_element(By.XPATH, '/html/body/div/form/button').click()
```

```python
#Preencher campos
#Nome/Razão social
navegador.find_element(By.ID, 'nome').send_keys('Danilo LTDA')
#Endereço
navegador.find_element(By.NAME, 'endereco').send_keys("Rua Engenheiro Willy Fischer")
#Bairro
navegador.find_element(By.NAME, 'bairro').send_keys("Vila Tiradentes")
#Municipio
navegador.find_element(By.NAME, 'municipio').send_keys("São Paulo")
#CEP
navegador.find_element(By.NAME, 'cep').send_keys("05368-050")
#UF
navegador.find_element(By.NAME, 'uf').send_keys("SP")
#CPF
navegador.find_element(By.NAME, 'cnpj').send_keys("123.456.789-00")
#IE
navegador.find_element(By.NAME, 'inscricao').send_keys("nao sei qual e")
#Descricao produtos
navegador.find_element(By.NAME, 'descricao').send_keys("Cursos para melhorar a vida")
#Quantidade
navegador.find_element(By.NAME, 'quantidade').send_keys("10")
#valor unitario
navegador.find_element(By.NAME, 'valor_unitario').send_keys(100)
#valor total
navegador.find_element(By.NAME, 'total').send_keys(1000)
```

```python
#Clicar em emitir nota fiscal
navegador.find_element(By.CLASS_NAME, 'registerbtn').click()
```
